# **⭐ Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as pd
```

```python
# import data
disease = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
# define y
y = disease['prognosis']
```

```python
# define X
X = disease.drop(['prognosis'],axis=1)
```

```python
# split data
from sklearn.model_selection import train_test_split
```

```python
# verify shape
X_train,X_test,y_train,y_test=train_test_split(X,y)
X_train.shape,X_test.shape,y_train.shape,y_test.shape

```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
# train model
model.fit(X_train,y_train)
```

```python
# predict with model
y_pred = model.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_percentage_error ,confusion_matrix , classification_report
```

```python
# model accuracy
mean_absolute_percentage_error(y_test,y_pred)
```

```python
# model confusion matrix
confusion_matrix(y_test,y_pred)
```

```python
# model classification report
print(classification_report(y_test,y_pred))
```
